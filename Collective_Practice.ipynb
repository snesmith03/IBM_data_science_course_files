```python
# 1. Take a string & count the number of upper, lower, numbers, & other characters 

string_mix = "Hi 2026!"
dict_chars = {}
# strings you can go through by character 
# (use the .get method where the key is in the brackets & as the key in the .get(key, default_value))
for n in string_mix:
    if n.isupper():
        dict_chars['upper'] = dict_chars.get('upper', 0) + 1
    elif n.islower():
        dict_chars['lower'] = dict_chars.get('lower', 0) + 1
    elif n.isdigit():
        dict_chars['digits'] = dict_chars.get('digits', 0) + 1
    else:
        dict_chars['special'] = dict_chars.get('special', 0) + 1

print(dict_chars)

'''
Note: You need to use the .isdigit(), .isupper(), .islower() methodds to see what the 
character of the element in the string is & note that a string IS an iterable, which you 
need for a for loop
'''
```

```python
# given a list, multiply even numbers by two and leave odd numbers alone 
lst_nums = [1, 2, 3, 4]
# define a new list 
new_lst = []
for num in lst_nums:
    if num % 2 == 0:
        new_lst.append(num * 2)
    else:
        new_lst.append(num)
print(new_lst)

'''
Note: You can change a list value in place but you have to reference the index position 
'''

nums_lst = [1, 3, 5, 6, 8]
for n in range(len(nums_lst)): # make sure to use len when you need to reference a list for range
    if nums_lst[n] % 2 == 0:
        nums_lst[n] = nums_lst[n] * 2
    else:
        pass # skips the element that does not meet some condition
print(nums_lst)
```

```python
# create a function that takes a list of numbers and returns the count of numbers > 50

def greater_fifty(lst_nums):
    count = 0 
    for num in lst_nums:
        if num > 50:
            count += 1
    return count 

# example 
print(greater_fifty([1, 51, 30, 45, 67]))
```

```python
# Write a function that takes a list of integers and returns a list of only the prime numbers, using a list comprehension

'''
List comprehension uses a list as with the loop and output embedded inside it 
'''

def primes_only(lst:list): # proper format type for specifying param ata type
    primes = []
    for l in lst:
        if l % 2 != 0:
            primes.append(l)
        else:
            pass
    return primes 
ex_1 = print(primes_only([1,2,3,4,5,6,7,8,9]))
```

```python
# Given a list of strings, return a dictionary mapping each string to the number of vowels it contains

lst_strings = ["hello", "world", "yesterday", "was", "a", "birthday"]
vowels = "a,e,i,o,u"
vowel_dict ={}
# use the .get function to count the elements associated with the vowel

# need an embedded for loop to look at the characters not the strings themselves 
for str in lst_strings:
    for i in str:
        if i in vowels and i not in vowel_dict: # if not in the dictionary
            vowel_dict[i] = vowel_dict.get(i, 0) + 1
        elif i in vowel_dict: # if already in the dictionary 
            vowel_dict[i] += 1
        else:
            pass
print(vowel_dict) 
```

```python
# You’re given a dictionary of student names → list of scores. Return a new dictionary mapping each student to their average score, but do not mutate the original.
students = {"Aisha": [88, 92, 79], "Marcus": [100, 94, 98, 96], "Lena": [72, 85], "Diego": [90, 90, 91, 89]}
student_avg = {}
for key, value in students.items(): # you need to reference the key-value pairs appropriately
    student_avg[key] = round((sum(value) / len(value)), 2)
print(student_avg)
```

```python
# Given a list of dicts with product and amount, compute total revenue per product.
transactions = [ {"product": "Laptop", "amount": 1200}, {"product": "Headphones", "amount": 150}, {"product": "Laptop", "amount": 1300}, {"product": "Monitor", "amount": 300}, {"product": "Headphones", "amount": 200}, {"product": "Keyboard", "amount": 90}, {"product": "Monitor", "amount": 250} ]

'''
Since the list holds the dictionaries you need to define the product and amount fields prior and then
you make an if-elif structure to specify what to do when something is or isn't in the dict

KEY: Think list -> dictionary elements

1. Loop over the list → each element is one dictionary.

2. Extract the specific keys you care about from that dictionary.

3. Use those extracted values to build whatever transformation or aggregation you need.
'''
product_dict = {}
for prod in transactions:
    product = prod["product"]
    amount = prod["amount"]
    if product not in product_dict:
        product_dict[product] = amount 
    elif product in product_dict:
        product_dict[product] += amount 

print(product_dict)
        
```
