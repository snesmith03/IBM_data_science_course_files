```python
# 1. Take a string & count the number of upper, lower, numbers, & other characters 

string_mix = "Hi 2026!"
dict_chars = {}
# strings you can go through by character 
# (use the .get method where the key is in the brackets & as the key in the .get(key, default_value))
for n in string_mix:
    if n.isupper():
        dict_chars['upper'] = dict_chars.get('upper', 0) + 1
    elif n.islower():
        dict_chars['lower'] = dict_chars.get('lower', 0) + 1
    elif n.isdigit():
        dict_chars['digits'] = dict_chars.get('digits', 0) + 1
    else:
        dict_chars['special'] = dict_chars.get('special', 0) + 1

print(dict_chars)

'''
Note: You need to use the .isdigit(), .isupper(), .islower() methodds to see what the 
character of the element in the string is & note that a string IS an iterable, which you 
need for a for loop
'''
```

```python
# given a list, multiply even numbers by two and leave odd numbers alone 
lst_nums = [1, 2, 3, 4]
# define a new list 
new_lst = []
for num in lst_nums:
    if num % 2 == 0:
        new_lst.append(num * 2)
    else:
        new_lst.append(num)
print(new_lst)

'''
Note: You can change a list value in place but you have to reference the index position 
'''

nums_lst = [1, 3, 5, 6, 8]
for n in range(len(nums_lst)): # make sure to use len when you need to reference a list for range
    if nums_lst[n] % 2 == 0:
        nums_lst[n] = nums_lst[n] * 2
    else:
        pass # skips the element that does not meet some condition
print(nums_lst)
```

```python
# write a function that takes a list of dictionaries representing students

def student_info(lst):
    max_score = 0
    for student in lst:
        if student["score"] > max_score:
            max_score = student["score"]
    # find the name 
    for student in lst:
        if student["score"] == max_score:
            print(student["name"])
    # you can use a list comprehension for min(), max(), sum(), etc. 
print(student_info([ {"name": "A", "score": 88}, {"name": "B", "score": 72}, {"name": "C", "score": 95} ]))
```

```python
# list comprehension - create a list of elements where word length > 3 lettters 

words = ["data", "science", "python", "AI"]
lst_greater_three = [word for word in words if len(word) > 3] # list comprehension 
print(lst_greater_three)

'''
Note:
Format for list comprehension: [iterable_var for iterable_var in iterable (some condition)]
'''
```
