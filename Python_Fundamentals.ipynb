Loops (for and while)

```python
# for loops 
cakes = ['vanilla', 'chocolate', 'strawberry', 'mint chocolate chip']
for cake in cakes:
    if 'e' in cake: # using the 'in' logic 
        print(cake)
    else:
        pass # skips the element 
```

```python
# while loops 

i = 0 # initialization 

while i <= 10: # condition 
    print(f'The value i holds is: {i}') # execution 
    i = round(i + 5 / 3, 3) # update 
```
