Lists 

```python
# lists are defined in brackets []
ice_cream = ['vanilla', 'chocolate', 'strawberry', 'mint chocolate']
print(ice_cream[2])
```

Tuples

```python
# tuples are a compound data structure and are defined in parentheses

tuple_1 = ('string', [1, 3, 4], (1, 2), 17.05, 19)
print(tuple_1[1][2]) # prints the first element in the list at the 1st index
```

Sets

```python
# sets are defined in brackets {}
list_one = ['red', 'orange', 'red', 'green', 'yellow', 'pink', 'purple', 'pink']
set_list_one = set(list_one)
print(set_list_one) # sets are unordered so they don't hold positions like lists or dictionaries or tuples
```

Dictionaries

```python
# dictionaries are formatted in key:value pairs 
# you can only access a value by referencing a key or with dict.values()

dict_flavors = {'lollipop': ['vanilla', 'strawberry'], 'ice cream': ['chocolate', 'rocky road', 'cotton candy'], 'jam': 'grape'}
print(dict_flavors['ice cream'][0]) # if you have lists in a dictionary you can reference specific indices
```
